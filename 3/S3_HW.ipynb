### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
delta_renovated_col_name = 'delta_renovated'
```

```python
df[['yr_renovated', 'yr_built']]
```

```python
#df.drop(columns=delta_renovated_col_name, inplace=True)
#df
```

```python
if delta_renovated_col_name not in df.columns:
    df[delta_renovated_col_name] = \
    df.apply(lambda x: x['yr_renovated']- x['yr_built']  
             if (x['yr_built']< x['yr_renovated'] and x['yr_renovated'] > 0) else None , axis=1)
else:
    df.drop(columns=delta_renovated_col_name, inplace=True)
df
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
month_col_name = 'date_month'
year_col_name = 'date_year'
```

```python
if month_col_name not in df.columns:
    df[month_col_name] = pd.to_datetime(df['date']).apply(lambda x: x.month)
else:
    ## df.drop(columns='date_month', inplace=True)
    pass
df.head()
```

```python
if year_col_name not in df.columns:
    df[year_col_name] = pd.to_datetime(df['date']).dt.year
else:
    df.drop(columns=year_col_name, inplace=True)
df
```

## 3 задача

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price':'mean'}).sort_values(by='bedrooms', ascending=False)
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('grade').agg({'price':['min', 'max', 'mean']}).sort_values(by='grade', ascending=True)
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.groupby(['waterfront','view']).agg({'id':'count'}).rename(columns={'id':'count'})
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
df.groupby(['floors','bedrooms']).agg({'id':'count'})\
.rename(columns={'id':'count'}).sort_values(by='count', ascending=False).head(1)
```
